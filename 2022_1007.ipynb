```python
# 3 (階差数列を第10項まで作成)
a_n = [0, 1, 3, 6, 10] # 初期数列設定(階差が等差)
m = 10 # 表示する項数

# 階差b_nの調査
b_n = [] # 階差
for i in range(len(a_n)-1):
  b_i = a_n[i+1] - a_n[i]
  b_n.append(b_i)
#print(b_n) # [1, 2, 3, 4]

# 階差b_nが等差数列の場合
if b_n[1]-b_n[0]==b_n[2]-b_n[1]:
  d = b_n[1]-b_n[0]
  b_i = b_n[len(a_n)-2]
  for i in range(m-len(a_n)):
    b_i = b_i + d
    b_n.append(b_i)
  #print(b_n)

# 数列a_n
a_i = a_n[-1]
for i in range(len(a_n) ,m):
  a_i = a_i + b_n[i-1]
  a_n.append(a_i)
print(a_n) # [0, 1, 3, 6, 10, 15, 21, 28, 36, 45]
```

```python
# 6 (フィボナッチ数列の第50項目を表示, 引数はf0, f1, n)
def fib(f0, f1, n): # n>=0
  if n==0:
    fib_list = [f0]
  else: # n>=1
    fib_list = [f0, f1]
    for i in range(n-1): # n=2のとき第2項となるよう調整 n>=2
      f2 = f0 + f1
      f0 = f1 # f0更新
      f1 = f2 # f1更新
      fib_list.append(f2)
  return fib_list

n = 50
f0 = 0
f1 = 1
fib_list = fib(f0, f1, n)
print("f0=", f0,", f1=", f1,"となるフィボナッチ数列の第", n,"項は", fib_list[n])
```

```python
# 8-(6) (動画をリサイズし連続表示)
import os

resize = 2 # リサイズ設定
# 動画読込
file_path = './data_1007/human_video.avi'
window_name = os.path.splitext(os.path.basename(file_path))[0] # ファイル名の部分だけ取得
cap_file = cv2.VideoCapture(file_path) # avi形式動画はmacでは再生できないようだ
print("動画読込確認:", cap_file.isOpened())

# 動画のプロパティを取得
width = int(cap_file.get(cv2.CAP_PROP_FRAME_WIDTH)) # 768.0
height = int(cap_file.get(cv2.CAP_PROP_FRAME_HEIGHT)) # 576.0
fps = int(cap_file.get(cv2.CAP_PROP_FPS)) # 25.0
print("幅:", width, ", 高さ:", height, ", FPS:", fps)
print("総フレーム数:", cap_file.get(cv2.CAP_PROP_FRAME_COUNT)) # 総フレーム数100.0 (4秒動画)

# 各フレームの読込
ret, frame = cap_file.read() # (読み込めたかどうか示すbool値, 画像の配列ndarray)のタプル
#print(ret) # Trueで読込できている
#print(type(frame)) # ndarray
#print(frame.shape) # (高さ:576, 幅:768, 色数:3)

# matplotlibによる画像確認
#frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
#plt.imshow(frame_rgb)
#plt.show()

# 動画保存用設定
fourcc = cv2.VideoWriter_fourcc('m','p','4','v') # 書き出し設定
writer = cv2.VideoWriter('./data_1007/human_video_resized.mp4', fourcc, fps, (int(width/resize), int(height/resize)))

# 画像をリサイズし, 連続表示
while True:
  ret, frame = cap_file.read()
  if not ret:
    break
  frame = cv2.resize(frame, (int(width/resize), int(height/resize))) # 1/2の大きさにリサイズ:(288, 384, 3)
  
  # openCVによる画像連続表示
  cv2.imshow(window_name, frame)
  cv2.waitKey(1)
  
  writer.write(frame) # 動画保存

# 画像ウィンドウを閉じる
cv2.destroyWindow(window_name)
cv2.waitKey(1)

writer.release()
cap_file.release()
```

4-3. 画像処理基礎

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# 1 (画像表示)
img2_bgr = cv2.imread('./data_1007/milkdrop.bmp')
img2_rgb = cv2.cvtColor(img2_bgr, cv2.COLOR_BGR2RGB)

# openCVによる画像表示
cv2.imshow('image', img2_bgr)
cv2.waitKey(1000) # 1秒表示
cv2.destroyAllWindows()
cv2.waitKey(1)

# matplotlibによる画像表示
plt.imshow(img2_rgb)
plt.show()
```

```python
# 2 (白色領域とそれ以外の2値画像を作成) 
img2_gray = cv2.imread('./data_1007/milkdrop.bmp', 0) # gray # cv2.cvtColor(img2_rgb, cv2.COLOR_RGB2GRAY)でも可
ret, thresh = cv2.threshold(img2_gray, 127, 255, cv2.THRESH_BINARY) # ret(retval)=127.0 # 127(閾値), 255(指定する色)
# メモ:閾値=127,122
#print(thresh) # 黒:0, 白:255 (0〜255) # thresh.shape=(256, 256)

# matplotlibによる画像表示
plt.imshow(thresh, 'gray') # gray
plt.title('BINARY')
plt.xticks([]) # x軸非表示
plt.yticks([]) # y軸非表示
plt.show()

# BGR画像保存
cv2.imwrite('./data_1007/milkdrop_binary.png', thresh)
```

```python
# 3 (白色領域の輪郭(outline)抽出)
img2_bgr = cv2.imread('./data_1007/milkdrop.bmp')

# 輪郭抽出
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 小さい輪郭は誤検出として削除
contours = list(filter(lambda x: cv2.contourArea(x) > 100, contours))
# メモ:len(contours)=2, len(contours[0])=389, len(hierarchy)=1, len(hierarchy[0])=37

# 面積が最大の輪郭を取得
contour = max(contours, key=lambda x: cv2.contourArea(x))

# 輪郭(outline)を作成
outline = cv2.drawContours(img2_bgr, [contour], -1, color=(255, 0, 0), thickness=2)
# drawContours : 輪郭の検出情報をもとに輪郭の描画を行う関数
# 第一引数 : 多次元配列(numpy.ndarray)
# 第二引数 : 輪郭を形成する画素(点)情報
# 第三引数 : 輪郭を形成する画素(点)のインデックス番号を指定する。例えば0を指定すると、1番目の輪郭を形成する画素(点)のみを描画する。1を指定すると、2番目の輪郭を形成する画素(点)のみを描画する。輪郭を形成する画素(点)を全て描画したい場合は、-1を指定する。
# 第四引数 : 輪郭を形成する画素(点)の色。BGR(Blue, Green, Red)指定
# 第五引数(任意) : 輪郭を形成する画素(点)の大きさを設定。デフォルト1
# 戻り値 : 多次元配列(numpy.ndarray)

# matplotlibによる画像表示
outline_rgb = cv2.cvtColor(outline, cv2.COLOR_BGR2RGB)
plt.imshow(outline_rgb)
plt.show()

# BGR画像保存
cv2.imwrite('./data_1007/milkdrop_outline.png', outline)
```

```python
# 4 (mask画像を作成)
# マスク画像を作成
mask = np.zeros_like(thresh) # np.zeros_lileではなく_likeが正しい
cv2.drawContours(mask, [contour], -1, color=255, thickness=-1)

# matplotlibによる画像表示
plt.imshow(mask, 'gray') # gray
plt.show()

# 画像保存
cv2.imwrite('./data_1007/milkdrop_mask.png', mask)
```

```python
# 5 (mask画像を用いてミルククラウン領域のみを表示)
# 画像再読込
img2_bgr = cv2.imread('./data_1007/milkdrop.bmp')
milk = img2_bgr

# mask画像を用いて指定領域以外削除
for j in range(len(milk)):
  for i in range(len(milk[j])):
    if mask[j][i]==0:
      milk[j][i] = [0, 0, 0]
#print(milk)

# matplotlibによる画像表示
milk_rgb = cv2.cvtColor(milk, cv2.COLOR_BGR2RGB)
plt.imshow(milk_rgb)
plt.show()

# BGR画像保存
cv2.imwrite('./data_1007/milkcrown_area.png', milk)
```
